## 🔧 Transformación
Aquí realizaremos la limpieza de datos, manejo de valores nulos y normalización de variables para preparar el dataset.

```python
# 1. Convertir 'Charges.Total' a numérico (suele venir como texto con espacios)
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce')

# 2. Manejo de valores nulos (Rellenar con 0 para este ejercicio)
df['Charges.Total'].fillna(0, inplace=True)

# 3. Mapear la variable objetivo 'Churn' a formato binario para Machine Learning
df['Churn_Binario'] = df['Churn'].map({'Yes': 1, 'No': 0})

# 4. Verificamos los cambios
df.info()
```

## 📊 Carga y análisis
Fase de Análisis Exploratorio de Datos (EDA) para identificar los motivos de la evasión de clientes.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Gráfico de Churn vs Tipo de Contrato
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='pastel')
plt.title('Tasa de Abandono (Churn) según el Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Cantidad de Clientes')
plt.show()

# (Opcional) Carga: Exportar el dataset limpio
# df.to_csv('telecom_x_clean.csv', index=False)
```
